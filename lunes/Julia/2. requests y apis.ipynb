## El módulo requests (hacer peticiones http)
* Incorporado directamente a partir de python3
* Permite acceder de forma sencilla a recursos externos basados en URL
* Proporciona métodos para lanzar peticiones HTTP mediante los métodos soportados por REST: get, post, put y delete

```python
import requests
response = requests.get('https://www.google.com')
```

```python
pars= {'key1': 'value1', 'key2': 'value2'}  # Se pueden enviar parámetros en la URL
response = requests.get('https://google.com/',params=pars)
print(response.url) #https://www.google.com/?key1=value1&key2=value2
response = requests.get('https://google.com/search?q=selva')
response.text
```

```python
# Las cabeceras (headers) de una petición a una API contienen información adicional que el servidor necesita para procesar la solicitud, pe autorizaciones y autentacaciones, formato de los datos, idioma, gestión de sesiones (cookies), idiomas y preferencias regionales, etc
heads= {'cabecera1': 'value1', 'cabecera2': 'value2'}
resp= requests.get('https://google.com/',headers=heads)
```

```python
resp.status_code
```

Códigos de estado
* 200: ok
* 404: recurso no encontrado
* 500: error interno del servidor
* 403: error de autenticación

```python
#  Acceder a la información de la respuesta: headers, json(), text
response = requests.get('https://google.com/')
for k, v in response.headers.items():
    print(k,"=",v)
```

```python
print(response.text)
```

```python
# API: (Interfaz de Programación de Aplicaciones)  conjunto de reglas y protocolos que permite que diferentes aplicaciones o sistemas intercambien información y se comuniquen entre sí. Funciona como un intermediario que facilita el acceso a funcionalidades o datos específicos de un servicio, plataforma o software
response = requests.get('https://swapi.dev/api/people/4')
sw=response.json()
sw
```

```python
#  Ejemplo para escribir ficheros
#  https://icdc.cen.uni-hamburg.de/thredds/catalog/ftpthredds/knsc/ocean/catalog.html
import requests
import os

url = 'https://icdc.cen.uni-hamburg.de/thredds/catalog/ftpthredds/knsc/ocean/annual/catalog.html?dataset=ftpthreddsscan/knsc/ocean/annual/KNSC_salinity__UHAM_ICDC__v2__0.5x0.25deg__1873.nc'
response = requests.get(url)

if response.ok:
        with open(os.path.basename(url), 'wb') as file_handler:
            file_handler.write(response.content)
            print('correcto')
else:
        print(f'Request failed: {response.text}')
```
